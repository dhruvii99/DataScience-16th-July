```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4])
labels = ["apple", "mango", "Cherry", "Dhruvi"]
myexplode = [0, 0.1, 0.2, 0.3]  # smaller explode values
c = ['red', 'green', 'blue', 'gray']

plt.pie(x, labels=labels, explode=myexplode, startangle=90, shadow=True, colors=c, autopct='%1.1f%%')
plt.show()
```
